```python

def step(x):
    if x >= 0:
        return 1
    else:
        return 0


def OR_neuron(x1, x2):
    w1 = 1
    w2 = 1
    b = -1
    return step(w1*x1 + w2*x2 + b)


def NAND_neuron(x1, x2):
    w1 = -1
    w2 = -1
    b = 1
    return step(w1*x1 + w2*x2 + b)


def AND_neuron(x1, x2):
    w1 = 1
    w2 = 1
    b = -2
    return step(w1*x1 + w2*x2 + b)


def XOR_network(x1, x2):
    h1 = OR_neuron(x1, x2)
    h2 = NAND_neuron(x1, x2)
    output = AND_neuron(h1, h2)
    return h1, h2, output

print("=========== XOR Neural Network ===========")
print("Architecture: (OR , NAND) -> AND")
print("  X1 |  X2 |  OR |  NAND | XOR ")

inputs = [(0,0), (0,1), (1,0), (1,1)]

for x1, x2 in inputs:
    h1, h2, y = XOR_network(x1, x2)
    print("  {}  |  {}  |  {}  |   {}   |  {}".format(x1, x2, h1, h2, y))

print("------------------------------------------")
```
