```python
def step(net):
    return 1 if net >= 0 else 0

X = [(0,0), (0,1), (1,0), (1,1)]

T = [0, 0, 0, 1]

w1, w2, b = 0, 0, 0

for x1, x2 in X:
    net = x1*w1 + x2*w2 + b
    y = step(net)
    t = T[X.index((x1,x2))]
    e = t - y

    w1 = w1 + e*x1
    w2 = w2 + e*x2
    b  = b  + e

    print(f"x=({x1},{x2}) y={y} e={e} -> w1={w1}, w2={w2}, b={b}")

print("\nFinal values:")
print("w1 =", w1, "w2 =", w2, "b =", b)
```

```python
import numpy as np

def pitts_neuron(x, w, theta):
    """McCulloch-Pitts neuron"""
    return 1 if np.dot(x, w) >= theta else 0

print("AND Gate")
w = np.array([1, 1])
theta = 2
for x in [(0,0), (0,1), (1,0), (1,1)]:
    print(pitts_neuron(np.array(x), w, theta))
```

```python
def step(net):
    return 1 if net >= 0 else 0

X = [(0,0), (0,1), (1,0), (1,1)]
T = [0, 0, 0, 1]

w1, w2, b = 0, 0, 0
epoch = 0

while True:
    epoch += 1
    error_count = 0

    for x1, x2 in X:
        net = x1*w1 + x2*w2 + b
        y = step(net)
        t = T[X.index((x1,x2))]
        e = t - y

        w1 = w1 + e*x1
        w2 = w2 + e*x2
        b  = b  + e

        print(f"Epoch {epoch}: x=({x1},{x2}) y={y} e={e} -> w1={w1}, w2={w2}, b={b}")

        if e != 0:
            error_count += 1

    if error_count == 0:
        break

print("\nFinal values after convergence:")
print("w1 =", w1, "w2 =", w2, "b =", b)
```
